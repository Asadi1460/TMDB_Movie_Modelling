# <a id='toc2_'></a>[<p style="background-color: #990033; font-family:calibri; color:white; font-size:150%; text-align:center; border-radius:20px 20px;">Dataset Descriptions and Data Understanding</p>](#toc0_)


- budget: The budget of the movie.
- genres: The genres of the movie.
- homepage: The homepage URL of the movie.
- id: Unique identifier for each movie.
- keywords: Keywords related to the movie.
- original_language: The original language of the movie.
- original_title: The original title of the movie.
- overview: A brief overview or summary of the movie.
- popularity: The popularity score of the movie.
- production_companies: Companies involved in producing the movie.
- production_countries: Countries where the movie was produced.
- release_date: The release date of the movie.
- revenue: The revenue generated by the movie.
- runtime: The runtime duration of the movie.
- spoken_languages: Languages spoken in the movie.
- status: The status of the movie (e.g., Released).
- tagline: The tagline of the movie.
- title: The title of the movie (again, duplicated).
- vote_average: The average vote/rating received by the movie.
- vote_count: The count of votes/ratings received by the movie.

```python
# Extract first genre
first_genre_names = []

for genres_json in df['genres']:
    genres_list = json.loads(genres_json)  # Parsing the JSON string into a list of dictionaries
    if genres_list:  # Checking if the list is not empty
        first_genre_name = genres_list[0]['name']  # Extracting the name of the first genre
        first_genre_names.append(first_genre_name)
    else:
        first_genre_names.append(None)  # Handling the case where genres list is empty

# Adding the first genre names as a new column to the DataFrame
df['first_genre'] = first_genre_names
```

```python
# Extract first production company
production_company = []

for production_companies_json in df['production_companies']:
    production_companies_list = json.loads(production_companies_json)  # Parsing the JSON string into a list of dictionaries
    if production_companies_list:  # Checking if the list is not empty
        first_production_company = production_companies_list[0]['name']  # Extracting the name of the first production company
        production_company.append(first_production_company)
    else:
        production_company.append(None)  # Handling the case where production_company list is empty

# Adding the production_company as a new column to the DataFrame
df['production_company'] = production_company
```

```python
df['production_company'].nunique()
```

```python
# Replace 'Columbia Pictures Corporation' with 'Columbia Pictures' in the 'production_company' column
df.loc[df['production_company'] == 'Columbia Pictures Corporation', 'production_company'] = 'Columbia Pictures'

# Replace 'Lionsgate' with 'Lions Gate Films' in the 'production_company' column
df.loc[df['production_company'] == 'Lionsgate', 'production_company'] = 'Lions Gate Films'
```

```python
top_production_companies = df['production_company'].value_counts().head(55) # only production_company are considerd that freq >= 10
top_production_companies
```

```python
top_production_company_list = top_production_companies.index.tolist()
top_production_company_list
```

```python
# Create a dictionary to store the count of movies for each production company
movies_per_company = {}

# Iterate over the top production companies list
for company in top_production_company_list:
    # Count the number of movies associated with the current production company
    count = df[df['production_company'] == company].shape[0]
    # Store the count in the dictionary
    movies_per_company[company] = count

# Print the dictionary containing the count of movies for each production company
movies_per_company
```

```python
# Calculate the total count of movies
total_count = sum(movies_per_company.values())

# Print the total count
print("Total count of movies:", total_count)
```

```python
top_production_companies_percent = round(total_count / df.shape[0] * 100, 2)
print('Percentage of Top Production Companies', top_production_companies_percent)
```

```python
# Filter the DataFrame to keep only rows where the production company is in the top_production_company_list
filtered_df = df[df['production_company'].isin(top_production_company_list)]

# Print the filtered DataFrame
filtered_df.shape
```

```python
def value_counts(df, column):
    # Calculate the frequency of each value in the "sex_of_casualty" column
    frequency = df[column].value_counts()
    
    # Print the frequency of every value
    print(f"Frequency of each value for DataFrame['{column}']: \n{frequency}")
```

```python
def remove_outliers(df, column, values_to_remove):
    """
    Remove rows from a DataFrame where the specified column contains specified values.

    Args:
        df (DataFrame): The pandas DataFrame.
        column (str): The column name to check for outliers.
        values_to_remove (list): A list of values to remove from the specified column.

    Returns:
        DataFrame: The DataFrame with outliers removed.
    """
    # Create a boolean mask to identify rows with the specified values to remove
    mask = df[column].isin(values_to_remove)
    
    # Use the mask to filter out rows with the specified values and create a new DataFrame
    filtered_df = df[~mask]
    
    # Optionally, you can modify the original DataFrame in place by uncommenting the following line:
    # df.drop(df[mask].index, inplace=True)
    
    return filtered_df
```

```python
def count_plot_df(df, column, label=None, orientation='v'):
    # Set the style of the seaborn plot
    sns.set_style("whitegrid")
    
    # Create the count plot
    plt.figure(figsize=(10, 6))
    if orientation == 'v':
        sns.countplot(data=df, x=column, palette="viridis")
        plt.xticks(rotation=90)  # Rotate x-axis labels for vertical orientation
    else:
        sns.countplot(data=df, y=column, palette="viridis")
    
    # Add labels and title
    if label is not None:
        plt.xlabel(label if orientation == 'v' else 'Frequency')
        plt.ylabel(label if orientation == 'h' else 'Frequency')
        plt.title(f"Frequency Plot of {label}")
    else:
        plt.title(f"Frequency Plot")

    # Show the plot
    plt.tight_layout()
    plt.show()
```

## <a id='toc20_6_'></a>[<span style='background-color:#990033; font-family:calibri; color:white; border-radius:15px 15px;'>ROC-AUC </span>](#toc0_)


```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Dictionary to store FPR and TPR for each classifier
roc_curves = {}

# Calculate ROC curve for each classifier
for name, model in models.items():
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_curves[name] = (fpr, tpr)

# Plot ROC curve for each classifier
plt.figure(figsize=(8, 6))
for name, (fpr, tpr) in roc_curves.items():
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], linestyle='--', color='grey', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```
